```python
import time
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense
import numpy
# fix random seed for reproducibility
numpy.random.seed(7)
```

```python
# load pima indians dataset
dataset = numpy.loadtxt("pima-indians-diabetes.data.csv", delimiter=",")
# split into input (X) and output (Y) variables
X = dataset[:,0:8]
Y = dataset[:,8]

# create model
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

start = time.clock()
# Fit the model
model.fit(X, Y, epochs=150, batch_size=10)
end = time.clock()

# evaluate the model
scores = model.evaluate(X, Y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
print(end)
print("time elapsed: %.2fs" % (end - start))
```

```python
from keras.utils import multi_gpu_model

# Replicates `model` on 2 GPUs.
# This assumes that your machine has 2 available GPUs.
parallel_model = multi_gpu_model(model, gpus=2)
parallel_model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

start = time.clock()
# This `fit` call will be distributed on 2 GPUs.
# Since the batch size is 256, each GPU will process 128 samples.
parallel_model.fit(X, Y, epochs=150, batch_size=10)
end = time.clock()

# evaluate the model
scores = parallel_model.evaluate(X, Y)
print("\n%s: %.2f%%" % (parallel_model.metrics_names[1], scores[1]*100))
print("time elapsed: %.2fs" % (end - start))
```
