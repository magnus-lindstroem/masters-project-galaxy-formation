```python
import time
import tensorflow as tf
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
import matplotlib.pyplot as plt
# fix random seed for reproducibility
np.random.seed(7)
```

```python
total_set_size = 10000
training_set_fraction = 0.8
validation_set_fraction = 0.1
test_set_fraction = 0.1

galfile = pd.read_hdf('/scratch/data/galcats/P200/galaxies.Z01.h5')
galaxies = galfile.as_matrix()
gal_header = galfile.keys().tolist()

print(gal_header)
print(galaxies[0])
print(galaxies.shape)
```

```python
mhalo = galaxies[:,6]
mpeak = galaxies[:,10]
mstar = galaxies[:,7]

print(mhalo[:10])
print(mstar[:10])

plt.scatter(mhalo, mstar, s=1)
```

```python
n_data_points = galaxies.shape[0]
subset_indices = np.random.choice(n_data_points, total_set_size, replace=False)
train_indices = subset_indices[:int(total_set_size*training_set_fraction)]
val_indices = subset_indices[int(total_set_size*training_set_fraction):
                                    int(total_set_size*(training_set_fraction+validation_set_fraction))]
test_indices = subset_indices[int(total_set_size*(training_set_fraction+validation_set_fraction)):]

mhalo_train = galaxies[train_indices, 6]
mhalo_val = galaxies[val_indices, 6]
mhalo_test = galaxies[test_indices, 6]

mstar_train = galaxies[train_indices, 7]
mstar_val = galaxies[val_indices, 7]
mstar_test = galaxies[test_indices, 7]
```

```python
# create model
model = Sequential()
model.add(Dense(8, input_dim=1, kernel_initializer='normal', activation='tanh'))
model.add(Dense(8, kernel_initializer='normal', activation='tanh'))
model.add(Dense(8, kernel_initializer='normal', activation='tanh'))
model.add(Dense(8, kernel_initializer='normal', activation='tanh'))
model.add(Dense(8, kernel_initializer='normal', activation='tanh'))
model.add(Dense(8, kernel_initializer='normal', activation='tanh'))
model.add(Dense(1, kernel_initializer='normal', activation=None))

# Compile model
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['mse'])

start = time.clock()
# Fit the model
history = model.fit(mhalo_train , mstar_train, validation_data=(mhalo_val, mstar_val), epochs=100000, batch_size=1000)
end = time.clock()

# evaluate the model
scores = model.evaluate(mhalo_test, mstar_test)
print('scores: ',scores)
print("\n%s: %.2f" % (model.metrics_names[1], scores[1]))
print(end)
print("time elapsed: %.2fs" % (end - start))
```

```python
predictedY = model.predict(mhalo_test)
error = model.evaluate(mhalo_test, mstar_test)
fig = plt.figure()
plt.scatter(mhalo_test, predictedY)
plt.scatter(mhalo_test, mstar_test, marker='.')
plt.xlabel('log($M_H/M_{sun}$)')
plt.ylabel('log($M_{gal}/M_{sun}$)')
plt.title('Backprop trained. Test set predictions. mse: %.5f.' % (error[1]))
fig.savefig('backprop_mhalo_to_mgalaxy.png')
```

```python
# list all data in history
print(history.history.keys())
```

```python
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper right')
plt.show()
```

```python
# summarize history for accuracy
plt.plot(history.history['mean_squared_error'])
plt.plot(history.history['val_mean_squared_error'])
plt.title('model accuracy')
plt.ylabel('mse')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
